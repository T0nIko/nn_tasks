```python
%matplotlib inline

import neurolab as nl
import numpy as np
import matplotlib.pylab as pl

pl.style.use('ggplot')
```

```python
n=50
# [1., 2.], [2., 1], [4., 3.], [3., 4.]
# 4 points will do either
xtrain = np.array([[1., 3.], [1.5, 1.5], [3., 1], [2., 4.], [3.5, 3.5], [4., 2.]])
x = np.random.uniform(0, 5, (n, 2))

pl.subplot(211)
pl.plot([0., 5.],[5., 0.], '-')
pl.plot(xtrain[:,0], xtrain[:,1], 'b+')
pl.ylabel("Train data")
pl.subplot(212)
pl.plot([0., 5.],[5., 0.], '-')
pl.plot(x[:,0], x[:,1], 'b+')
pl.ylabel("Generated data")
pl.show()

d = np.sign(xtrain[:, 0] + xtrain[:, 1] - 5.).reshape(6, 1)
```

```python
net = nl.net.newff([[0, 5],[0, 5]], [1, 1])

err = net.train(xtrain, d, epochs=1000, show=100, goal=0.0001)

out = net.sim(x)

pl.plot([0., 5.],[5., 0.], 'b')

for i in range(n):
    if out[i]>0:
        pl.plot(x[i,0], x[i,1], 'ro')
        
    if out[i]<0:
        pl.plot(x[i,0], x[i,1], 'go')
        
pl.show()
```

# Amount of points

6 point selections show pretty high accuracy, if those training point are distribued equally between 2 clusters. Though, some datasets might fail a bit, especially if point distribution not that random.
