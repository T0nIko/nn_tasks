```python
%matplotlib inline

import neurolab as nl
import numpy as np
import matplotlib.pylab as pl

pl.style.use('ggplot')
```

```python
n = 50

lim_ls = np.linspace(0, 5, 50)
y = lim_ls**2 - 4*lim_ls + 5

y = [i if i <= 5 else 5 for i in y]

x = np.random.uniform(0, 5, (n, 2))
xtrain = np.random.uniform(0, 5, (n, 2))

d = np.zeros(n)

for i in range(n):
    if xtrain[i, 1] - xtrain[i, 0]**2 + 4*xtrain[i, 0] > 5:
        d[i] = 1
    else:
        d[i] = -1
        
pl.subplot(211)
pl.plot(lim_ls, y, 'brown')
pl.ylabel("Train data")
pl.plot(xtrain[:,0], xtrain[:,1], 'b*')
pl.subplot(212)
pl.plot(lim_ls, y, 'brown')
pl.ylabel("Generated data")
pl.plot(x[:,0], x[:,1], 'b+')
pl.show()
```

```python
net = nl.net.newff([[0, 5], [0, 5]],[8, 1])

error = net.train(xtrain, d.reshape(n, 1), epochs=1000, show=10, goal=0.0001)

out = net.sim(x)

for i in range(n):
    if out[i] > 0:
        pl.plot(x[i,0], x[i,1], 'ro')
    else:
        pl.plot(x[i,0], x[i,1], 'go')

pl.plot(lim_ls, y, 'brown')
pl.ylabel('Net output')
pl.show()
```
