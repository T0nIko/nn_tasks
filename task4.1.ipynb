```python
%matplotlib inline

import neurolab as nl
import numpy as np
import matplotlib.pylab as pl

pl.style.use('ggplot')
```

```python
n=4

center = np.array([[0.2, 0.2], [0.4, 0.4], [0.7, 0.3], [0.2, 0.5]])

k=100

x = np.zeros(n*k*2)

x = x.reshape(n*k, 2)

for i in range(n):
    rand_norm = 0.05 * np.random.randn(k, 2)
    for j in range(k):
        x[i*k+j] = center[i] + rand_norm[j]
    
np.random.shuffle(x)
```

```python
net = nl.net.newc([[0.0, 1.0],[0.0, 1.0]], 20)

error = net.train(x, epochs=200, show=20)
```

```python
w = net.layers[0].np['w']

pl.plot(center[:,0], center[:, 1], 'bo', x[:,0], x[:,1], 'g+', w[:,0], w[:,1], 'ro')
pl.legend(['real centers','train samples', 'train centers'])

pl.show()
```
